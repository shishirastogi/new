```python
import matplotlib.pyplot as plt

# Data: 12 months of sales
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 
          'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
sales = [12000, 15000, 11000, 18000, 22000, 20000, 
         25000, 24000, 28000, 32000, 35000, 40000]

# Create the plot
plt.figure(figsize=(10, 5))
plt.plot(months, sales, marker='o', color='tab:blue', linewidth=2, label='Monthly Sales')

# Add labels and title
plt.title('Company Sales Performance (12 Months)', fontsize=14)
plt.xlabel('Month', fontsize=12)
plt.ylabel('Sales Revenue ($)', fontsize=12)

# Aesthetic improvements
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()

# Display the plot
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data for 6 months
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']

# Sales data for three products
sales_a = [1500, 1800, 1600, 2100, 1900, 2400]
sales_b = [1200, 1300, 1700, 1600, 2100, 2300]
sales_c = [2000, 1900, 2200, 2500, 2300, 2800]

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(months, sales_a, marker='o', label='Product A', linewidth=2)
plt.plot(months, sales_b, marker='s', label='Product B', linewidth=2)
plt.plot(months, sales_c, marker='^', label='Product C', linewidth=2)

# Customization
plt.title('Monthly Sales for Products A, B, and C (6 Months)', fontsize=14)
plt.xlabel('Month', fontsize=12)
plt.ylabel('Sales ($)', fontsize=12)
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(42)
marks = np.random.randint(30, 100, 100)

plt.figure(figsize=(10, 6))
n, bins, patches = plt.hist(marks, bins=10, edgecolor='black')

cm = plt.cm.get_cmap('tab10')
for i in range(len(patches)):
    patches[i].set_facecolor(cm(i % 10))

plt.title('Distribution of Marks Obtained by 100 Students', fontsize=14)
plt.xlabel('Marks', fontsize=12)
plt.ylabel('Number of Students', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(42)
hours_studied = np.random.uniform(1, 10, 50)
marks_obtained = 20 + 7 * hours_studied + np.random.normal(0, 5, 50)
marks_obtained = np.clip(marks_obtained, 0, 100)

plt.figure(figsize=(10, 6))
plt.scatter(hours_studied, marks_obtained, color='purple', alpha=0.6, edgecolors='k')
plt.title('Relationship Between Hours Studied and Marks Obtained', fontsize=14)
plt.xlabel('Hours Studied', fontsize=12)
plt.ylabel('Marks Obtained', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

cities = ['Tokyo', 'New York', 'London', 'Mumbai', 'Paris']
population = [37.4, 8.4, 9.0, 20.4, 2.1]
area = [2194, 783, 1572, 603, 105]
gdp = [1600, 1700, 600, 310, 810]

plt.figure(figsize=(10, 6))
bubble_sizes = [g * 2 for g in gdp]
plt.scatter(population, area, s=bubble_sizes, alpha=0.5, c=np.arange(len(cities)), cmap='viridis', edgecolors='k')

for i, txt in enumerate(cities):
    plt.annotate(txt, (population[i], area[i]), fontsize=10, ha='center', va='center')

plt.title('City Population vs Area (Bubble size: GDP)', fontsize=14)
plt.xlabel('Population (Millions)', fontsize=12)
plt.ylabel('Area (sq km)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

```python
import matplotlib.pyplot as plt

years = [2018, 2019, 2020, 2021, 2022, 2023]
profits = [50000, 65000, 45000, 80000, 95000, 110000]

plt.figure(figsize=(10, 6))
plt.fill_between(years, profits, color="skyblue", alpha=0.4)
plt.plot(years, profits, color="Slateblue", alpha=0.6, linewidth=2)

plt.title('Yearly Profit Growth', fontsize=14)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Profit ($)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.5)

plt.show()
```

```python
import matplotlib.pyplot as plt

hours = [0, 4, 8, 12, 16, 20, 24]
consumption = [0.5, 0.4, 1.2, 2.5, 1.8, 3.5, 0.7]

plt.figure(figsize=(10, 6))
plt.step(hours, consumption, where='post', color='teal', linewidth=2, marker='o')

plt.title('Daily Electricity Consumption Step Chart', fontsize=14)
plt.xlabel('Time of Day (24-hour format)', fontsize=12)
plt.ylabel('Consumption (kW)', fontsize=12)
plt.xticks(hours)
plt.grid(True, linestyle='--', alpha=0.5)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(42)
eng = np.random.normal(80000, 15000, 50)
sales = np.random.normal(70000, 12000, 50)
hr = np.random.normal(60000, 10000, 50)

plt.figure(figsize=(10, 6))
plt.boxplot([eng, sales, hr], labels=['Engineering', 'Sales', 'HR'], patch_artist=True,
            boxprops=dict(facecolor='lightblue'))
plt.title('Salary Distribution by Department', fontsize=14)
plt.ylabel('Salary ($)', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

np.random.seed(42)
class_a = np.random.normal(70, 10, 100)
class_b = np.random.normal(80, 5, 100)
class_c = np.random.normal(65, 15, 100)

df = pd.DataFrame({'Class A': class_a, 'Class B': class_b, 'Class C': class_c})
df_melted = df.melt(var_name='Class', value_name='Score')

plt.figure(figsize=(10, 6))
sns.violinplot(x='Class', y='Score', data=df_melted, palette='muted')
plt.title('Exam Score Distribution by Class', fontsize=14)
plt.xlabel('Class', fontsize=12)
plt.ylabel('Exam Score', fontsize=12)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
sales = [10, 25, 18, 35, 30, 45]
categories = ['A', 'B', 'C', 'D']
values = [15, 30, 45, 10]
brands = ['Brand X', 'Brand Y', 'Brand Z', 'Others']
market_share = [40, 30, 20, 10]
np.random.seed(42)
x_scatter = np.random.rand(50)
y_scatter = np.random.rand(50)

fig, axs = plt.subplots(2, 2, figsize=(12, 10))

axs[0, 0].plot(months, sales, marker='o', color='b')
axs[0, 0].set_title('Line Chart')

axs[0, 1].bar(categories, values, color='orange')
axs[0, 1].set_title('Bar Chart')

axs[1, 0].pie(market_share, labels=brands, autopct='%1.1f%%', startangle=90)
axs[1, 0].set_title('Pie Chart')

axs[1, 1].scatter(x_scatter, y_scatter, color='purple', alpha=0.6)
axs[1, 1].set_title('Scatter Plot')

plt.tight_layout()
plt.show()
```
